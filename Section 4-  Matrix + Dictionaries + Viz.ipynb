```python
# Not aesthetically pleasing, because of the decimal places, so we round
round (FieldGoals/Games)
```

```python
# WE get the error, since for matrix we need to use a different round function, as below:
np.matrix.round(FieldGoals/Games)
```

```python
# For this division, we need both the matrices need to have the same dimensions
# nan is for 0/0
FieldGoalsPerGame= np.matrix.round(FieldGoals/Games)
```
