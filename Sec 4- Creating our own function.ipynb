# Creating Our Own Function 

```python
def myplot (playerlist):
    for name in playerlist:
        plt.plot(Games[Pdict[name]],c='Black',ls="--",marker='s',label=name)
    
    plt.legend(loc='upper left', bbox_to_anchor=(1,1)) 
    plt.xticks(list(range(0,10)),Seasons, rotation='vertical')

    plt.show() 
# the iterative variable is player, which iterates over the values in player list
# once we call the function, we pass a list of player's name
# based on that list, the pdict function gives index to games, and graph for particular player is plotted
```

```python
myplot(["KobeBryant","LeBronJames","DerrickRose"])
```
