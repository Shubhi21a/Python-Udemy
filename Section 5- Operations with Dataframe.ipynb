```python
stats[Filter2]
```

```python
stats[stats.BirthRate>40] #using the filter directly
```

```python
stats[BirthRate>40] #This doesn't work , it has to be stats[stats.BirthRate>40]
```

```python
stats[stats.BirthRate>40 and stats.InternetUsers<2] #in R, we wouldn't get an error
# python is not vectorized 
# this 'and' doesn't know how to work with objects/lists
# we want it to check values for each row and then give us result, but it expects ONE true/false value in all, just 1
# and that is not possible
```

```python
stats[stats.BirthRate>40 & stats.InternetUsers<2] # doesn't know the order, we add (), to make it clear
```

```python
stats[(stats.BirthRate>40) & (stats.InternetUsers<2)] # Done! :)
```

# Visualizing with Seaborn

```python
vis3=sns.lmplot(data=stats, x='InternetUsers', y='BirthRate')
# on shift+tab, you'll see that the syntax says we specify x, y first, and then data; but we can also do it as above
# first specifying the dataset, and getting that out of way, then x,y; like in R
# lm- linear model, regression model
```

```python
vis3=sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', fit_reg=False)
#clicked shift+tab to see the syntax; we remove the regression line, since the data points aren't really linear fitting
```

```python
vis3=sns.lmplot(data=stats, x='InternetUsers',y='BirthRate',fit_reg=False,hue='IncomeGroup')
#keyword for colour= hue, by default hue=None
```

```python
vis3=sns.lmplot(data=stats, x='InternetUsers',y='BirthRate',fit_reg=False,hue='IncomeGroup',size=10, aspect=2)
#keyword for colour= hue, by default hue=None
#size, coze this follows other rules, and hence our command earlier doesn't increase its size
#aspect helps adjust the lenght* breadth 
```

```python
vis3=sns.lmplot(data=stats, x='InternetUsers',y='BirthRate',fit_reg=False,hue='IncomeGroup',size=10, aspect=0.5)
#keyword for colour= hue, by default hue=None
#size, coze this follows other rules, and hence our command earlier doesn't increase its size
#aspect helps adjust the lenght* breadth 
```

```python
vis3=sns.lmplot(data=stats, x='InternetUsers',y='BirthRate',fit_reg=False,hue='IncomeGroup',size=10, aspect=1)
#keyword for colour= hue, by default hue=None
#size, coze this follows other rules, and hence our command earlier doesn't increase its size
#aspect helps adjust the lenght* breadth 
```
